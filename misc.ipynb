```python
import glob
class_label = {0:'ignored',1:'pedestrian',2:'people',3:'bicycle',4:'car',5:'van',6:'truck',7:'tricycle',8:'awning-tricycle',9:'bus',10:'motor',11:'others'}
val_files = glob.glob('val_ann\*.txt')
for file in val_files[:]:
    with open(file, 'r') as f:
        with open(file.replace('val_ann','val_ann_new'),'w') as f2:
            for l in f.readlines():
                left,top,width,height,_,c,_,_ = l.split(',')
                right =int(left)+int(width)
                bottom = int(top) +int(height)
                message = f'{class_label[int(c)]} {left} {top} {right} {bottom}'
                f2.write(f'{message}\n')
                print(message)
```
